```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

import seaborn as sns

sns.set_theme(style="whitegrid", palette="muted")
```

```python
def f(x):
    return x**2

def df(x):
    return 2*x
```

```python
# f の x における接戦をプロット
def tangent_plot(f, df, x, x_range=(-10, 10), label="f(x)", figsize=(4, 6)):    
    fig = plt.figure(figsize=figsize)
    _x = np.linspace(*x_range, 100)
    plt.plot(_x, f(_x), label=label)
    plt.plot(_x, f(x) + df(x) * (_x - x))
    plt.plot(x, f(x), "o", label=f"x={x}")
    plt.legend()
```

```python
tangent_plot(f, df, 2.9, label="f(x) = x^2")
plt.savefig("../img/fx2_tangent_x2.9.png")
```

```python
f = lambda x: (x**2) / 10 + 10 * np.sin(x**2 / 4)
df = lambda x: x / 5 + 5 * np.cos(x**2 / 4)
plt.figure(figsize=(10, 6))
plt.plot(
    np.linspace(-7, 7, 100),
    f(np.linspace(-7, 7, 100)),
    label="f(x) = x^2 / 10 + 10 * sin(x^2 / 4)",
)
plt.legend()
plt.savefig("../img/f_sin.png")
```

```python
tangent_plot(f, df, 1.5, label="f(x) = x^2 / 10 + 10 * sin(x^2 / 4)", figsize=(10,6), x_range=(-7, 7))
plt.ylim(-10, 20)
plt.savefig("../img/fx2_tangent_sin.png")
```

```python
x0 = 3
eta = 0.1

def gradient_descent(f, df, x0, eta, n_iter=100001):
    history = [x0]
    x = x0
    for _ in range(n_iter):
        x = x - eta * df(x)
        history.append(x)
    return history
```

```python
gradient_descent(f, df, x0, eta, n_iter=101)
```

```python
f = lambda x: x**2 + np.exp(-x)
df = lambda x: 2*x - np.exp(-x)

gradient_descent(f, df, x0, eta=0.01, n_iter=1000)
```
